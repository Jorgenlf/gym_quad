### To generate the profiler file open the terminal and run 

``` python -m cProfile -o profiling/profiler_result.prof  -s cumulative (-m module | myscript.py) ```

(from the gym_quad directory)

- -o writes the profile results to a file instead of to stdout

- -s specifies one of the sort_stats() sort values to sort the output by. This only applies when -o is not supplied. See image below.

- -m specifies that a module is being profiled instead of a script.

### To visualize
Run 

``` snakeviz profiling/profiler_result.prof ```

in a terminal from the gym_quad directory.

#### This will open an interactive webpage for the visualization 
You might have to reduce depth or cutoff to be able to see the visualization

### To only view the printout, run the cell below

````python
import pstats 
stats = pstats.Stats('profiler_result.prof')
stats.strip_dirs().sort_stats(-1).print_stats()
````
